```python
# To standardize the 'Gender' column in the dataset
employee_data['Gender'] = employee_data['Gender'].replace(['M', 'male'], 'Male')
employee_data['Gender'] = employee_data['Gender'].replace(['F', 'female'], 'Female')

employee_data 
```

```python
# To know how many times each unique value appears in 'Date of Birth' column 
employee_data['Date of Birth'].value_counts()
```

```python
# converting the DOB column to datetime
employee_data['Date of Birth'] = pd.to_datetime(employee_data['Date of Birth'],
    errors='coerce', dayfirst=False)

# Replacing invalid date in DOB column
employee_data['Date of Birth'] =employee_data['Date of Birth'].replace({'30-02-1993' : '1993-02-28'})

# filling NaN values with the replace date
employee_data['Date of Birth'] = employee_data['Date of Birth'].fillna('1993-02-28')

employee_data['Date of Birth']
```

```python
employee_data['Date of Birth'].describe()
```

```python
# To check the unique items in the 'Department' column
employee_data['Department'].unique()
```

```python
# trying to work with a duplicate of the dataset
employee_data = employee_data.copy()

# To standardize the 'Department' column
Dept_data = employee_data['Department']
Dept_data ={'R&D' : 'Research & Development', 'HR' : 'Human Resources', 'engineering' : 'Engineering'}

# To replace the Department column with the standardize items
employee_data['Department'] = employee_data['Department'].replace(Dept_data)

employee_data['Department']
```

```python
employee_data['Department'].value_counts()
```

```python
# To check the unique items in 'Salary' column
employee_data['Salary'].unique()
```

```python
# trying to work with a duplicate of the dataset
employee_data = employee_data.copy()

# Convert valid numeric values temporarily for calculation
numeric_part = pd.to_numeric(employee_data['Salary'], errors='coerce')

# calculating min and mode to fill the rows with 'abc' & 'not disclosed'
min_salary = int(numeric_part.min())
mode_salary = int(numeric_part.mode()[0])

# Replace 'abc' with min_salary and 'not disclosed' with mode_salary
employee_data.loc[employee_data['Salary'] == 'abc', 'Salary'] = min_salary
employee_data.loc[employee_data['Salary'] == 'not disclosed', 'Salary'] = mode_salary

# Convert to int safely
employee_data['Salary'] = pd.to_numeric(employee_data['Salary']).astype('int64')

employee_data['Salary']
```

```python
# To know how many times each unique value appears in 'Join Date' column 
employee_data['Join Date'].value_counts()
```

```python
# Converting Join Date column to datetime, handling both slashes and dashes, and invalid dates
employee_data['Join Date'] = pd.to_datetime(employee_data['Join Date'],
    errors='coerce', dayfirst=True )

# Convert the placeholder to datetime
default_date = pd.to_datetime('2000-01-01')

# Fill missing values with default_date
employee_data['Join Date'] = employee_data['Join Date'].fillna(default_date)

employee_data['Join Date']
```

```python
employee_data['Email'].value_counts()
```

```python
# trying to work with a duplicate of the dataset
employee_data = employee_data.copy()

# converting datatype in 'Email' column to string & remove all other @ after the first one
employee_data['Email'] = employee_data['Email'].str.replace(r'@(.*)@', r'@\1', regex=True)
employee_data['Email']
```

# Exploratory Data Analysis

```python
# To count the numbers of Male & Female in each Gender
gender_counts = employee_data['Gender'].value_counts()
gender_counts
```

```python
# To calculate the average salary for each Gender & Group by Gender
avg_salary_by_gender = employee_data.groupby('Gender')['Salary'].mean().sort_values(ascending=False)

avg_salary_by_gender
```

```python
# To count the number of employees in each department
department_counts = employee_data['Department'].value_counts().sort_values(ascending=True)

department_counts
```

```python
#  # To calculate the average salary for each department & Group by department
avg_salary_by_dept = employee_data.groupby('Department')['Salary'].mean().sort_values(ascending=True)

avg_salary_by_dept
```

```python
# To know the numbers of Male & Female Gender in each department
dept_headcount = employee_data.groupby('Department')['Gender'].value_counts()
dept_headcount
```

```python
# Defining Salary Range into low, medium & high
def categorize_salary(salary):
    if salary < 50000:
        return 'Low'
    elif 50000 <= salary < 70000:
        return 'Medium'
    else:
        return 'High'
        
# Applying the define Salary Range
employee_data['Salary Range'] = employee_data['Salary'].apply(categorize_salary)

# To count how many employees fall into each Range
employee_data['Salary Range'].value_counts()
```

```python
# To create Age column & Calculate age
Age = pd.to_datetime('today').year
employee_data['Age'] = Age - employee_data['Date of Birth'].dt.year

# To convert from float to int
employee_data['Age'] = employee_data['Age'].astype('int64')

employee_data['Age']
```

```python
# Defining custom bins and labels for Age Range
bins = [20, 30, 40, 50, 60, 70]
labels = ['20â€“29', '30â€“39', '40â€“49', '50â€“59', '60â€“69']

# Create a new column with age groups
employee_data['Age Range'] = pd.cut(employee_data['Age'], bins=bins, labels=labels, right=False)
employee_data['Age Range']
```

```python
# Count how many employees are in each age group
age_range_counts = employee_data['Age Range'].value_counts().sort_index()
age_range_counts 
```

```python
# Calculate average salary per age group
avg_salary_by_age = employee_data.groupby('Age Range', observed=False)['Salary'].mean().sort_values(ascending=False)
avg_salary_by_age
```

```python
employee_data[['Age', 'Salary']].describe()
```

```python
# Extracting the year people were hired
employee_data['Join Year'] = employee_data['Join Date'].dt.year

# Count hires per year
hires_count_per_year = employee_data['Join Year'].value_counts().sort_index()
hires_count_per_year
```

```python
# To know how many numbers each dept hires every year
hires_per_year = employee_data.groupby('Join Year')['Department'].value_counts()
hires_per_year
```

```python
# To create Years of Experience column & to Calculate employee Years of Experience
employee_data['Experience'] = pd.to_datetime('today').year - employee_data['Join Date'].dt.year
employee_data['Experience']
```

# Visualization

```python
# Bar plot for gender count
gender_counts.plot(kind='bar', color=['steelblue', 'goldenrod'], figsize=(5, 5))
plt.title('Gender Distribution', size = 16)
plt.xlabel('Gender', size = 14)
plt.ylabel('Count', size = 18)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Barh plot for department count
department_counts.plot(kind='barh', color=['lightcoral', 'goldenrod', 
        'darkorange', 'green', 'purple', 'darkcyan', 'steelblue'])
plt.title('Number of Employees per Department')
plt.xlabel('Count', size = 14)
plt.ylabel('Department', size = 16)
plt.tight_layout()
plt.show()
```

```python
# To know the number of employees that falls within a particular Range
age_range_counts.plot(kind='bar', color=['green', 'steelblue', 'purple', 'darkcyan', 'darkorange'])
plt.title('Employees Age Ranges', size = 17) 
plt.xlabel('Age Range', size = 14)
plt.ylabel('Number of Employees', size = 16)
plt.xticks(rotation=0)
plt.show()
```

```python
# histogram plot for salary range: low, medium, high
sns.histplot(employee_data['Salary Range'], bins=30, kde=True, color='goldenrod')
plt.title('Salary Distribution', size = 18)
plt.xlabel('Salary Ranges', size = 14)
plt.ylabel('Number of Employees', size = 16)
plt.tight_layout()
plt.show()
```

```python
# Line chart to show the number of people hires every year
hires_count_per_year.plot(kind='line', marker='o', markersize=10, markeredgecolor='goldenrod', 
                    markerfacecolor='goldenrod', color='darkcyan')
plt.title('Hiring Trend by Year', size = 18)
plt.xlabel('Year', size = 16)
plt.ylabel('Number of Hires', size = 17)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Bar chart to show the average salary rate accross department
avg_salary_by_dept.plot(kind='bar', color=['lightcoral', 'goldenrod', 
    'darkorange', 'green', 'purple', 'darkcyan', 'steelblue'])
plt.title('Average Salary by Department', size = 14)
plt.xlabel('Department', size = 16)
plt.ylabel('Average Salary', size = 14)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# scatterplot chart to show the correlation between Salary and years of experience
sns.scatterplot(data=employee_data, x='Experience', y='Salary', color='darkorange', s=50)
plt.title('Salary vs. Years of Experience', size = 14)
plt.xlabel('Years of Experience', size = 12)
plt.ylabel('Salary', size = 16)
plt.grid(axis = 'x', color = 'darkcyan', linewidth= 1.5)
plt.tight_layout()
plt.show()
```

```python
sns.lmplot(data=employee_data, x='Experience', y='Salary', line_kws={'color' : 'darkorange'})
plt.title('Trend: Salary vs. Experience', size = 14)
plt.xlabel('Years of Experience', size = 12)
plt.ylabel('Salary', size = 16)
plt.tight_layout()
plt.show()
```

```python
# To check for outliers
sns.boxplot(x=employee_data['Age'], color='orchid')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.tight_layout()
plt.show()
```

```python
# To check for outliers
sns.boxplot(x=employee_data['Salary'], color='teal')
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.tight_layout()
plt.show()
```
